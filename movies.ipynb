```python
def make_work_easy(obj):
    l1 = []
    for i in ast.literal_eval(obj):
        l1.append(i['name'])
    return l1
```

```python
movies['genres']=movies['genres'].apply(make_work_easy) #made an function for getting values of keys which is genres
```

```python
def make_work_easy2(obj):
    l1 = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l1.append(i['name'])
            counter +=1
        else:
            break
    return l1
```

```python
movies['cast'] = movies['cast'].apply(make_work_easy2) #made on function for getting values of keys which is cast
```

```python
def direc(obj):
    l1 = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l1.append(i['name'])
            break
    return l1
```

```python
movies['crew'] = movies['crew'].apply(direc) #made on function for getting values of keys which is director
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower()) #making all content of tags column to lower case
```

```python
from sklearn.feature_extraction.text import CountVectorizer #importing Counter Vectorizer to vertorize our data
cv = CountVectorizer(max_features=5000,stop_words='english') 
```

```python
vectors = cv.fit_transform(df['tags']).toarray() #fitting the content to the model
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving') #for getting root words we are using ps.stem
```

```python
def convert(text):
    t = []
    for i in text.split():
        t.append(ps.stem(i))
    return " ".join(t) 
```

```python
df['tags'] = df['tags'].apply(convert) #made on function for getting values of the words in the tags column
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors) #for finding distance we are using cosine function
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #in this we are first finding out top 5 most similars movies then sorting them on the similaity rank
```

```python
def recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title) #getting recommended name of movies
```

```python
recommend('The Dark Knight')
```
